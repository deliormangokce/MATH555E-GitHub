```python
import numpy as np
import matplotlib.pyplot as plt

a=0
b=5
N=10
dx = (b-a)/N
def f(x):
    return x**2
xi=0
sum =0
for i in range(N):
    xi = xi+dx;
    sum = sum + f(xi)
print(sum*dx)

```
