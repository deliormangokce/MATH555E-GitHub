## 2. Education datataset

This dataset is taken from [Kaggle](https://www.kaggle.com/spscientist/students-performance-in-exams). For this dataset, construct a decision tree regression and a XGBoost regression model to predict

1. math score
2. reading score
3. writing score

Beware: you might have to transform the data first before you train and test your models.

```python
education = pd.read_csv("https://raw.githubusercontent.com/deliormangokce/MATH555E-GitHub/main/StudentsPerformance.csv")
education
```

```python
education.info()
education.describe()
education['parental level of education'].describe()
```

```python
X1=education['gender'].map({'female': 1, 'male': 2})
X2=education['race/ethnicity'].map({'group A': 1, 'group B': 2,'group C': 3,'group D': 4,'group E': 5})
X3=education['parental level of education'].map({"associate's degree": 1, 'some college': 2,'high school': 3,"master's degree": 4,"bachelor's degree": 5, 'some high school': 6})
X4=education['lunch'].map({'standard': 1, 'free/reduced': 2})
X5=education['test preparation course'].map({'completed': 1, 'none': 2})
Y1=education['math score']
Y2=education['reading score']
Y3=education['writing score']

data=(X1,X2,X3,X4,X5)
X=pd.DataFrame(data)
X=X.transpose()
X
```

```python
##for Y1
xtrain,xtest,ytrain,ytest = train_test_split(X,Y1,test_size = 0.25, random_state = 0)
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(xtrain, ytrain)
```

```python
y1_pred = regressor.predict(xtest)
```

```python
## xgb model for y1
xgb = xgboost.XGBRegressor(n_estimators=100, learning_rate=0.08, gamma=0, subsample=0.75,colsample_bytree=1, max_depth=7)
xgb.fit(xtrain,ytrain)
```

```python
predictions = xgb.predict(xtest)
pd.DataFrame({'Actual':ytest, 'XGBoost regr_predict':predictions, 'DecisionTree_predict':y1_pred})
```

```python
##for Y2
xtrain,xtest,ytrain,ytest = train_test_split(X,Y2,test_size = 0.25, random_state = 0)
regressor = DecisionTreeRegressor()
regressor.fit(xtrain, ytrain)
y2_pred = regressor.predict(xtest)
```
